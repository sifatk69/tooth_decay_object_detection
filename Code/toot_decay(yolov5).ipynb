# Split Training and Validation dataset

```python
#arrays to store file names
imgs =[]
txts =[]

#setup dir names
trainPath = '/content/dataset/yolov5/Dataset/images/train'
valPath = '/content/dataset/yolov5/Dataset/images/valid'
crsPath = '/content/dataset/yolov5/data/augmented_images_yolo5_3/' #dir where images and annotations stored

#setup ratio (val ratio = rest of the files in origin dir after splitting into train and test)
train_ratio = 0.8
val_ratio = 0.2


#total count of imgs
totalImgCount = len(os.listdir(crsPath))/2

#soring files to corresponding arrays
for (dirname, dirs, files) in os.walk(crsPath):
    for filename in files:
        if filename.endswith('.txt'):
            txts.append(filename)
        else:
            imgs.append(filename)


#counting range for cycles
countForTrain = int(len(imgs)*train_ratio)
countForVal = int(len(imgs)*val_ratio)
print("training images are : ",countForTrain)
print("Validation images are : ",countForVal)
```

```python
os.makedirs("/content/dataset/yolov5/Dataset/images/train")
```

```python
os.makedirs("/content/dataset/yolov5/Dataset/labels/train")
```

```python
trainimagePath = '/content/dataset/yolov5/Dataset/images/train'
trainlabelPath = '/content/dataset/yolov5/Dataset/labels/train'

#cycle for train dir
for x in range(countForTrain):

    fileJpg = choice(imgs) # get name of random image from origin dir
    fileTxt = fileJpg[:-4] +'.txt' # get name of corresponding annotation file

    #move both files into train dir
    shutil.move(os.path.join(crsPath, fileJpg), os.path.join(trainimagePath, fileJpg))
    shutil.move(os.path.join(crsPath, fileTxt), os.path.join(trainlabelPath, fileTxt))


    #remove files from arrays
    imgs.remove(fileJpg)
    txts.remove(fileTxt)



```
